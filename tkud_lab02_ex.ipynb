## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
import numpy as np
arr = np.random.randint(0,1001,100)
print("10 phan tu dau tien: ", arr[:10])
print("5 phan tu cuoi cung: ", arr[-5:])
print("Cac phan tu chia het cho 5 va lon hon 100: ", arr[(arr % 5==0) & (arr > 100)])
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
print("cac phan tu be hon 100: ", arr[arr < 100])
print("so luong cac phan tu be hon 100: ", len(arr[arr < 100]))
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
print("Tong mang: ", np.sum(arr))
print("Gia tri trung binh: ", np.mean(arr))
print("Gia tri trung vi: ", np.median(arr))
print("Q1: ", np.percentile(arr, 25))
print("Q2: ", np.percentile(arr, 50))
print("Q3: ", np.percentile(arr, 75))
print("Phan vi 20: ", np.percentile(arr, 20))
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
# Chọn ngẫu nhiên 10 phần tử từ mảng không thay thế, thực hiện 5 lần
random_selections = [np.random.choice(arr, size=10, replace=False) for _ in range(5)]

for i, selection in enumerate(random_selections, start=1):
    print(f"Lần {i}: {selection}")
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
# Tìm phần tử xuất hiện nhiều nhất
unique_elements, counts = np.unique(arr, return_counts=True)
most_frequent_element = unique_elements[np.argmax(counts)]
frequency = counts.max()
print(f"Phần tử xuất hiện nhiều nhất: {most_frequent_element} (xuất hiện {frequency} lần)")
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
x = np.linspace(0, 2 * np.pi, 500)
sin_x = np.sin(x)
cos_x = np.cos(x)

plt.plot(x, sin_x, label="sin(x)", color="blue")

# Đồ thị cos(x)
plt.plot(x, cos_x, label="cos(x)", color="orange" , linestyle='--')

plt.title("Đồ thị sin(x) và cos(x)", fontsize=14)

# Thêm chú thích và lưới
plt.legend()
plt.grid(True)

plt.show()
```

### Câu 8: Vẽ histogram

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ histogram các số ngẫu nhiên trên.

```python
# This cell contains output. Don't remove!
```

```python
# Phát sinh 10,000 số ngẫu nhiên từ phân phối chuẩn
random_numbers = np.random.normal(size=10000)

# Vẽ histogram
plt.figure(figsize=(8, 6))
plt.hist(random_numbers, bins=50, color='skyblue', edgecolor='black', alpha=0.7)

# Thêm tiêu đề và nhãn
plt.title("Histogram của 10,000 số ngẫu nhiên từ phân phối chuẩn", fontsize=14)
plt.xlabel("Giá trị", fontsize=12)
plt.ylabel("Tần suất", fontsize=12)

# Hiển thị biểu đồ
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### Câu 9: Vẽ đồ thị hộp

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ boxplot các số ngẫu nhiên trên.

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
# Phát sinh 10,000 số ngẫu nhiên từ phân phối chuẩn
random_numbers = np.random.normal(size=10000)

# Vẽ boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(random_numbers, patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'))

# Thêm tiêu đề và nhãn
plt.title("Boxplot của 10,000 số ngẫu nhiên từ phân phối chuẩn", fontsize=14)
plt.ylabel("Giá trị", fontsize=12)

# Hiển thị biểu đồ
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

```python
# This cell contains output. Don't remove!
```

```python
# Phát sinh 10,000 số ngẫu nhiên từ phân phối chuẩn
random_numbers = np.random.normal(size=10000)

# Vẽ boxplot với phân phối nằm ngang
plt.figure(figsize=(8, 6))
plt.boxplot(random_numbers, patch_artist=True, vert=False, boxprops=dict(facecolor='skyblue', color='black'))

# Thêm tiêu đề và nhãn
plt.title("Boxplot ngang của 10,000 số ngẫu nhiên từ phân phối chuẩn", fontsize=14)
plt.xlabel("Giá trị", fontsize=12)

# Hiển thị biểu đồ
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
