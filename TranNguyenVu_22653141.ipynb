```python
import pandas as pd 
import numpy as np
data = pd.read_csv('data/thoi_tiet_khong_khi_vn_2023.csv')

# Hiển thị 10 dòng đầu tiên  
data.head(10)
```

```python
#10 dòng cuối cùng của bảng dữ liệu.
data.tail(10)
```

```python
# Cho biết bảng dữ liệu có bao nhiêu dòng và bao nhiêu cột?
data.shape
# 3650 dong va 18 cot
```

```python
# Cho biết thông tin về các cột dữ liệu và kiểu của từng cột trong bảng.
data.info()
```

```python
# Cho biết số lượng bản ghi bị thiếu giá trị cho các cột: Nhiệt_độ, Lượng_mưa và AQI.
print('số lượng bản ghi bị thiếu giá trị cho cột Nhiệt_độ: ', data['Nhiệt_độ'].isnull().sum())
print('số lượng bản ghi bị thiếu giá trị cho cột Lượng_mưa:  ', data['Lượng_mưa'].isnull().sum())
print('số lượng bản ghi bị thiếu giá trị cho cột AQI:  ', data['AQI'].isnull().sum())
```

```python
#Với cột Nhiệt_độ, hãy tính các trị thống kê:
 
print('Trung bình: ', data['Nhiệt_độ'].mean())
print('Trung vị: ', data['Nhiệt_độ'].median())
print('Độ lệch chuẩn: ', data['Nhiệt_độ'].std())
mode_1 = data['Nhiệt_độ'].mode()
print('Mode: ', mode_1.tolist())
print('PHương sai: ', data['Nhiệt_độ'].var())
print('Nhiệt độ cao nhất: ', data['Nhiệt_độ'].max())
print('Nhiệt độ thấp nhất: ', data['Nhiệt_độ'].min())
print('biên độ nhiệt: ', data['Nhiệt_độ'].max() - data['Nhiệt_độ'].min())
```

```python
# Dựa vào dữ liệu cột Thời_tiết, anh/chị hãy thực hiện:
# Đếm số ngày có thời tiết "Nắng" và số ngày có thời tiết "Mưa to" hoặc "Dông".
print('Số ngày có thời thiết nắng ', data[data['Thời_tiết'] == "Nắng"].shape[0])
print('Số ngày có thời thiết Mưa to hoặc dông ', data[data['Thời_tiết'] == "Mưa to"].shape[0] | data[data['Thời_tiết'] == "Dông"].shape[0])
```

```python
# Vẽ đồ thị Pie để so sánh phân bố các loại thời tiết trong năm.

data['Thời_tiết'].value_counts().plot(kind='pie', autopct='%.2f%%')
```

```python
# Liệt kê 20 ngày có chỉ số AQI cao nhất và thành phố tương ứng.


data[["AQI", "Thành_phố"]].sort_values(by='AQI', ascending=False).head(20)
```

```python
#2.	(0.75đ) Tính nhiệt độ trung bình theo tháng cho mỗi thành phố và hiển thị kết quả.
data.groupby(['Thành_phố'])["Nhiệt_độ"].mean()
```

```python
#3.	(0.25đ) Cho biết thành phố có lượng mưa trung bình năm cao nhất.
data.groupby(['Thành_phố'])["Lượng_mưa"].max().head(1)
```

```python
# Thống kê để hoàn thành bảng số liệu dưới đây.
nhietdo = data.groupby(['Khu_vực'])["Nhiệt_độ"].mean()
luongmua = data.groupby(['Khu_vực'])["Lượng_mưa"].mean()
aqi = data.groupby(['Khu_vực'])["AQI"].mean()

   
#df
#2.	(0.75đ) Từ đó trực quan hóa dữ liệu Bảng 1 bằng đồ thị Bar.
```

```python
df = pd.DataFrame(columns=['Khu vực', 'Nhiệt độ TB', 'Lượng mưa TB', 'AQI TB'])
df['Khu vực'] = nhietdo.index
df['Nhiệt độ TB'] = nhietdo.values
df['Lượng mưa TB'] = luongmua.values
df['AQI TB'] = aqi.values
df

```

```python
#2.	(0.75đ) Từ đó trực quan hóa dữ liệu Bảng 1 bằng đồ thị Bar.
import matplotlib.pyplot as plt
df.plot(kind='bar', x='Khu vực', y='Nhiệt độ TB')
plt.show()
```
